```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('historic_groceries2.csv')
options = ['Mar','Sep']
df2 = df[(df['Month'].isin(options))]
df2
```

```python
sns.set(style = 'whitegrid')
sns.stripplot(x="Month", y="Rice", data=df2)
  
plt.show()
```
