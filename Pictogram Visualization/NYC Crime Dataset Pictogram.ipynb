```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('NYPD_Arrest_Data__Year_to_Date_.csv')
df
```

Grabbing number of perperators per race and overall number of perperators.

```python
ucount = 0
AAcount = 0
APcount = 0
BHcount = 0
WHcount = 0
Wcount = 0
Bcount = 0
ovRcount = 0
for obj in df['PERP_RACE']:
    ovRcount += 1
    if obj == 'UNKNOWN':
        ucount += 1
    if obj == 'AMERICAN INDIAN/ALASKAN NATIVE':
        AAcount += 1
    if obj == 'ASIAN / PACIFIC ISLANDER':
        APcount += 1
    if obj == 'BLACK HISPANIC':
        BHcount += 1
    if obj == 'WHITE HISPANIC':
        WHcount += 1
    if obj == 'WHITE':
        Wcount += 1
    if obj == 'BLACK':
        Bcount += 1
```

```python
print('Proportion of unknown perperators to overall number ' + str(ucount/ovRcount))
print('Proportion of american indian/alaskan native perperators to overall number ' + str(AAcount/ovRcount))
print('Proportion of asian/pacific islander perperators to overall number ' + str(APcount/ovRcount))
print('Proportion of black hispanic perperators to overall number ' + str(BHcount/ovRcount))
print('Proportion of white hispanic perperators to overall number ' + str(WHcount/ovRcount))
print('Proportion of white perperators to overall number ' + str(Wcount/ovRcount))
print('Proportion of black perperators to overall number ' + str(Bcount/ovRcount))
```
