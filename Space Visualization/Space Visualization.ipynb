```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.graph_objs as go
```

```python
df2 = pd.read_csv('Space_Corrected2.csv')
df2.sort_values('Year',inplace=True)
df2.head(50)
```

```python
df = df2.assign(n=0).groupby(['Year', 'Rocket'])['n'].count().reset_index()
df = df[df["Rocket"] < 1600]
v = df.pivot(index='Rocket', columns='Year', values='n').fillna(0).values.tolist()
```

```python
iplot([go.Surface(z=v)])
```
