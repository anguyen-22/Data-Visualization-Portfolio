```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
afm_data = np.loadtxt('./afm.txt')
print(afm_data[0:5])
```

```python
afm_data *= (10**9)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)

ax.xaxis.set_tick_params(size=0)
ax.yaxis.set_tick_params(size=0)
ax.set_xticks([])
ax.set_yticks([])

ax.fill_between(x=[1.4, 1.9], y1=[0.1, 0.1], y2=[0.2, 0.2], color='black')
ax.text(x=1.65, y=0.25, s='500 nm', va='bottom', ha='center', color='black', size=20)

img = ax.imshow(afm_data, origin='lower', cmap='Spectral_r', extent=(0, 2, 0, 2), vmin=0, vmax=200)
```
